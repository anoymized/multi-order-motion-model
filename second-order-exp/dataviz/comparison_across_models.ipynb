```python
def load_model_data(file_rootx, file_rooty):
    dfx = pd.read_csv(file_rootx, encoding='utf-8')
    dfy = pd.read_csv(file_rooty, encoding='utf-8')
    Mod1_model_X = dfx['type1'].values
    Mod2_model_X = dfx['type2'].values
    Mod3_model_X = dfx['type3'].values
    Mod4_model_X = dfx['type4'].values
    Mod5_model_X = dfx['type5'].values
    Mod6_model_X = dfx['type6'].values
    Mod7_model_X = dfx['type7'].values

    Mod1_model_Y = dfy['type1'].values
    Mod2_model_Y = dfy['type2'].values
    Mod3_model_Y = dfy['type3'].values
    Mod4_model_Y = dfy['type4'].values
    Mod5_model_Y = dfy['type5'].values
    Mod6_model_Y = dfy['type6'].values
    Mod7_model_Y = dfy['type7'].values

    # to numpy
    Mod1_model_X, Mod1_model_Y = np.array(Mod1_model_X), np.array(Mod1_model_Y)
    Mod2_model_X, Mod2_model_Y = np.array(Mod2_model_X), np.array(Mod2_model_Y)
    Mod3_model_X, Mod3_model_Y = np.array(Mod3_model_X), np.array(Mod3_model_Y)
    Mod4_model_X, Mod4_model_Y = np.array(Mod4_model_X), np.array(Mod4_model_Y)
    Mod5_model_X, Mod5_model_Y = np.array(Mod5_model_X), np.array(Mod5_model_Y)
    Mod6_model_X, Mod6_model_Y = np.array(Mod6_model_X), np.array(Mod6_model_Y)
    Mod7_model_X, Mod7_model_Y = np.array(Mod7_model_X), np.array(Mod7_model_Y)

    # cartesian to polar
    Mod1_model_a, Mod1_model_v = cartesian2polar(Mod1_model_X, Mod1_model_Y)
    Mod2_model_a, Mod2_model_v = cartesian2polar(Mod2_model_X, Mod2_model_Y)
    Mod3_model_a, Mod3_model_v = cartesian2polar(Mod3_model_X, Mod3_model_Y)
    Mod4_model_a, Mod4_model_v = cartesian2polar(Mod4_model_X, Mod4_model_Y)
    Mod5_model_a, Mod5_model_v = cartesian2polar(Mod5_model_X, Mod5_model_Y)
    Mod6_model_a, Mod6_model_v = cartesian2polar(Mod6_model_X, Mod6_model_Y)
    Mod7_model_a, Mod7_model_v = cartesian2polar(Mod7_model_X, Mod7_model_Y)

    Mod1_model_a, Mod2_model_a, Mod3_model_a, Mod4_model_a, Mod5_model_a, Mod6_model_a, Mod7_model_a, \
        = np.rad2deg(Mod1_model_a), np.rad2deg(Mod2_model_a), \
        np.rad2deg(Mod3_model_a), np.rad2deg(Mod4_model_a), np.rad2deg(Mod5_model_a), np.rad2deg(Mod6_model_a), \
        np.rad2deg(Mod7_model_a),

    Mod1_model_XY = np.concatenate((Mod1_model_X, Mod1_model_Y))
    Mod2_model_XY = np.concatenate((Mod2_model_X, Mod2_model_Y))
    Mod3_model_XY = np.concatenate((Mod3_model_X, Mod3_model_Y))
    Mod4_model_XY = np.concatenate((Mod4_model_X, Mod4_model_Y))
    Mod5_model_XY = np.concatenate((Mod5_model_X, Mod5_model_Y))
    Mod6_model_XY = np.concatenate((Mod6_model_X, Mod6_model_Y))
    Mod7_model_XY = np.concatenate((Mod7_model_X, Mod7_model_Y))

    return Mod1_model_XY, Mod2_model_XY, Mod3_model_XY, Mod4_model_XY, Mod5_model_XY, Mod6_model_XY, Mod7_model_XY, \
              Mod1_model_a, Mod2_model_a, Mod3_model_a, Mod4_model_a, Mod5_model_a, Mod6_model_a, Mod7_model_a, \
                Mod1_model_v, Mod2_model_v, Mod3_model_v, Mod4_model_v, Mod5_model_v, Mod6_model_v, Mod7_model_v

```
